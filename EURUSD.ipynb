```python
#Functions to fetch the maxima and minima and fit the patterns

def get_max_min(prices, smoothing, window_range):
    smooth_prices = prices['Close'].rolling(window=smoothing).mean().dropna()
    local_max = argrelextrema(smooth_prices.values, np.greater)[0]
    local_min = argrelextrema(smooth_prices.values, np.less)[0]
    price_local_max_dt = []
    for i in local_max:
        if (i>window_range) and (i<len(prices)-window_range):
            price_local_max_dt.append(prices.iloc[i-window_range:i+window_range]['Close'].idxmax())
    price_local_min_dt = []
    for i in local_min:
        if (i>window_range) and (i<len(prices)-window_range):
            price_local_min_dt.append(prices.iloc[i-window_range:i+window_range]['Close'].idxmin())  
    maxima = pd.DataFrame(prices.loc[price_local_max_dt])
    minima = pd.DataFrame(prices.loc[price_local_min_dt])
    max_min = pd.concat([maxima, minima]).sort_index()
    max_min.index.name = 'date'
    max_min = max_min.reset_index()
    max_min = max_min[~max_min.date.duplicated()]
    p = prices.reset_index()   
    max_min['day_num'] = p[p['Time'].isin(max_min.date)].index.values
    max_min = max_min.set_index('day_num')['Close']
    
    return max_min

def find_IHS(max_min):  
    patterns = []
    # Window range is 5 units
    for i in range(5, len(max_min)):  
        window = max_min.iloc[i-5:i]
        # Pattern must play out in less than n units
        #if window.index[-1] - window.index[0] >n:      
         #   continue   
        a, b, c, d, e = window.iloc[0:5]
        if a<b and c<a and c<e and c<d and e<d and abs(b-d)<=np.mean([b,d])*0.02:
            patterns.append([window.index[-1], d])   
    return patterns

def find_SYMTRIBULL(max_min):  
    patterns = []
    # Window range is 6 units
    for i in range(6, len(max_min)):  
        window = max_min.iloc[i-6:i]
        # Pattern must play out in less than n units
        #if window.index[-1] - window.index[0] <n:      
         #   continue   
        a, b, c, d, e, f = window.iloc[0:6]
        if a>b and b<c and c>d  and d<e and e>f and b<d<f and a>c>e:
            patterns.append([window.index[-1], e])  
    return patterns

def find_ASCTRI(max_min):  
    patterns = []
    # Window range is 6 units
    for i in range(6, len(max_min)):  
        window = max_min.iloc[i-6:i]
        # Pattern must play out in less than n units
        #if window.index[-1] - window.index[0] <n:      
         #   continue   
        a, b, c, d, e, f = window.iloc[0:6]
        if a>b and b<c and c>d  and d<e and e>f and b<d<f and max(abs(a-c), abs(a-e),abs(c-e))<=np.mean([a,c,e])*0.02:
            patterns.append([window.index[-1], e])  
    return patterns

def find_BULLFLAG(max_min):  
    patterns = []
    # Window range is 6 units
        
    for i in range(6, len(max_min)): 
        window = max_min.iloc[i-6:i]
        # Pattern must play out in less than n units
        #if window.index[-1] - window.index[0] <n:      
         #   continue   
        a, b, c, d,e,f = window.iloc[0:6]
        if a>b and b<c and c>d and d<e and e>f and b>d>f and a>c>e:
            patterns.append([window.index[-1], e])  
    return patterns

def find_DOUBLEBOTTOM(max_min):  
    patterns = []
    # Window range is 3 units
    for i in range(3, len(max_min)):  
        window = max_min.iloc[i-3:i]
        # Pattern must play out in less than n units
       # if window.index[-1] - window.index[0] <n:      
        #    continue   
        a, b, c = window.iloc[0:3]
        if a<b and b>c and abs(a-c)<=np.mean([a,c])*0.02:
            patterns.append([window.index[-1], b]) 
    return patterns

def find_TRIPLEBOTTOM(max_min):  
    patterns = []
    # Window range is 5 units
    for i in range(5, len(max_min)):  
        window = max_min.iloc[i-5:i]
        # Pattern must play out in less than n units
        #if window.index[-1] - window.index[0] <n:      
         #   continue   
        a, b, c, d, e = window.iloc[0:5]
        if a<b and b>c and c<d and d>e and max(abs(a-c), abs(a-e), abs(e-c))<=np.mean([a,c,e])*0.02 and abs(b-d)<=np.mean([b,d])*0.02:
            patterns.append([window.index[-1], d]) 
    return patterns

def find_BULLWEDGE(max_min):  
    patterns = []
    # Window range is 6 units
    for i in range(6, len(max_min)):  
        window = max_min.iloc[i-6:i]
        # Pattern must play out in less than n units
        #if window.index[-1] - window.index[0] <n:      
         #   continue   
        a, b, c, d,e,f = window.iloc[0:6]
        if a>b and b<c and c>d and d<e and e>f and a<c<e and b<d<f and f-b>e-a:
            patterns.append([window.index[-1], e]) 
    return patterns

def find_HS(max_min):  
    patterns = []
    # Window range is 5 units
    for i in range(5, len(max_min)):  
        window = max_min.iloc[i-5:i]
        # Pattern must play out in less than n units
       # if window.index[-1] - window.index[0] <n:      
        #    continue   
        a, b, c, d, e = window.iloc[0:5]
        if a>b and c>a and c>e and c>d and e>d and abs(b-d)<=np.mean([b,d])*0.02:
            patterns.append([window.index[-1], d]) 
    return patterns

def find_SYMTRIBEAR(max_min):  
    patterns = []
    # Window range is 6 units
    for i in range(6, len(max_min)):  
        window = max_min.iloc[i-6:i]
        # Pattern must play out in less than n units
      #  if window.index[-1] - window.index[0] <n:      
       #     continue   
        a, b, c, d, e, f= window.iloc[0:6]
        if a<b and b>c and c<d and d>e and e<f and  b>d>f and a<c<e:
            patterns.append([window.index[-1], e]) 
    return patterns

def find_DESCTRI(max_min):  
    patterns = []
    # Window range is 6 units
    for i in range(6, len(max_min)):  
        window = max_min.iloc[i-6:i]
        # Pattern must play out in less than n units
       # if window.index[-1] - window.index[0] <n:      
        #    continue   
        a, b, c, d, e, f = window.iloc[0:6]
        if a<b and b>c and c<d and d>e and e<f and b>d>f and max(abs(a-c),abs(a-e), abs(c-e))<=np.mean([a,c,e])*0.02:
            patterns.append([window.index[-1], e]) 
    return patterns

def find_BEARFLAG(max_min):  
    patterns = []
    # Window range is 6 units
    for i in range(6, len(max_min)):  
        window = max_min.iloc[i-6:i]
        # Pattern must play out in less than n units
       # if window.index[-1] - window.index[0] <n:      
        #    continue   
        a, b, c, d,e,f = window.iloc[0:6]
        if a<b and b>c and c<d and d>e and e<f and a<c<e and b<d<f:
            patterns.append([window.index[-1], e])  
    return patterns

def find_DOUBLETOP(max_min):  
    patterns = []
    # Window range is 3 units
    for i in range(3, len(max_min)):  
        window = max_min.iloc[i-3:i]
        # Pattern must play out in less than n units
       # if window.index[-1] - window.index[0] <n:      
        #    continue   
        a, b, c = window.iloc[0:3]
        if a>b and b<c and abs(a-c)<=np.mean([a,c])*0.02:
            patterns.append([window.index[-1], b]) 
    return patterns

def find_TRIPLETOP(max_min):  
    patterns = []
    # Window range is 5 units
    for i in range(5, len(max_min)):  
        window = max_min.iloc[i-5:i]
        # Pattern must play out in less than n units
       # if window.index[-1] - window.index[0] <n:      
        #    continue   
        a, b, c, d, e = window.iloc[0:5]
        if a>b and b<c and c>d and d<e and max(abs(a-c), abs(a-e), abs(e-c))<=np.mean([a,c,e])*0.02 and abs(b-d)<=np.mean([b,d])*0.02:
            patterns.append([window.index[-1], d]) 
    return patterns

def find_BEARWEDGE(max_min):  
    patterns = []
    # Window range is 6 units
    for i in range(6, len(max_min)):  
        window = max_min.iloc[i-6:i]
        # Pattern must play out in less than n units
       # if window.index[-1] - window.index[0] <n:      
        #    continue   
        a, b, c, d,e,f = window.iloc[0:6]
        if a<b and b>c and c<d and d>e and e<f and a>c>e and b>d>f and b-f>a-e:
            patterns.append([window.index[-1], e]) 
    return patterns

def prev_max(data,extrema,n):
    data=data.Close
    points=extrema.index
    LN=[0]*(len(data))
    window=15
    for i in range(n-1,len(points)-n-window):
        LN[points[i]+window:points[i+n]+window]=[extrema[points[i-n+1]]]*(points[i+n]-points[i])
    for i in range(0,len(LN)):
        LN[i]=100*(data[i]-LN[i])/data[i]
    return LN

def prev_max1(data,extrema,n):
    data=data.Close
    points=extrema.index
    LN=[0]*(len(data))
    window=0
    for i in range(n-1,len(points)-n-window):
        LN[points[i]+window:points[i+n]+window]=[extrema[points[i-n+1]]]*(points[i+n]-points[i])
    for i in range(0,len(LN)):
        LN[i]=100*(data[i]-LN[i])/data[i]
    return LN

def signal_converter(date, entry, data):
    data=data.Close
    signals=[0]*len(data)
    for j in range(0,len(date)):
        k=min(60,len(data)-date[j])
        for i in range(0,k):
            if  data[date[j]]<entry[j] and data[date[j]+i]>entry[j]:
                signals[date[j]+i]=1
                break
            if  data[date[j]]>entry[j] and data[date[j]+i]<entry[j]:
                signals[date[j]+i]=1
                break
    return signals
```
